# Data Augmentation

- Data augmentation is a good technique we can refer when there is less quality imaeges
- it will help to process low resolution or quality images
- we can show the same image with difference way by using the Data Augmenation
-image.png

```python
from tensorflow.keras.utils import img_to_array , load_img
```

```python
img = load_img('/content/elephant.jpg')
```

```python
img.show()
```

```python
# convert image to the numerical array
x = img_to_array(img)
```

```python
x.shape    # pixel values will return
x = x.reshape((1, ) + x.shape)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

```

```python
datagen =  ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    rescale = 1./255 ,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
i = 0
for batch in datagen.flow(x,batch_size=1,save_to_dir='/content',save_prefix='ele',save_format='jpeg'):
    i += 1
    if i > 20:
        break
```
