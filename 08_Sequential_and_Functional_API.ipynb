```python
import tensorflow as tf
from tensorflow.keras.utils import plot_model
```

## Sequential API

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu' , input_shape=(2,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation = 'sigmoid')
])


```

```python
plot_model(model, show_shapes=True)
```

```python
# we can do sequetial API like that

model1 = tf.keras.Sequential()

model1.add(tf.keras.layers.Dense(64, activation='relu' , input_shape=(2,)))
model1.add(tf.keras.layers.Dense(16, activation='relu'))
model1.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
plot_model(model1, show_shapes=True)
```

## Functional API

```python
input = tf.keras.layers.Input(shape=(2,))

hidden1 = tf.keras.layers.Dense(64, activation='relu')(input)
hidden2 = tf.keras.layers.Dense(16, activation='relu')(hidden1)

output = tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)

model2 = tf.keras.Model(inputs=[input], outputs=[output])
```

```python
plot_model(model2, show_shapes=True)
```

- In here we have two input layers and one output layer

image.png

```python
# let us create the left side of the layers

input1 = tf.keras.layers.Input(shape=(2,))
norm1 = tf.keras.layers.Normalization()(input1)
hidden1 = tf.keras.layers.Dense(16,activation='relu')(norm1)

# let us create the right side of the layers

input2 = tf.keras.layers.Input(shape=(2,))
hidden2 = tf.keras.layers.Dense(32,activation='relu')(input2)

concat = tf.keras.layers.Concatenate()([hidden1,hidden2])

output = tf.keras.layers.Dense(1,activation='sigmoid')(concat)

model5 =tf.keras.Model(inputs=[input1,input2],outputs=output)
```

```python
plot_model(model5,show_shapes=True,show_layer_names=True)
```

- in here we are going to create a single input and multiple outputs

```python
input =  tf.keras.layers.Input(shape=(2,))

hidden1 = tf.keras.layers.Dense(16,activation='relu')(input)
hidden2 = tf.keras.layers.Dense(32,activation='relu')(hidden1)

output1 = tf.keras.layers.Dense(1,activation='sigmoid')(hidden2)
output2 = tf.keras.layers.Dense(3,activation='sigmoid')(hidden2)

model8 = tf.keras.Model(inputs=input,outputs=[output1,output2])
```

```python
plot_model(model8,show_shapes=True)
```
