```python
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import opendatasets as od
```

```python
od.download("https://www.kaggle.com/datasets/rupakroy/sonarcsv")
```

```python
#load data set
df = pd.read_csv('/content/sonarcsv/sonar.csv')
df.head()
```

```python
# Divide input and the output
x = df.iloc[:,:60]
y = df.iloc[:,60]
```

```python
#Encode Lables
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y[:4]
```

```python
# Scale Features
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
# split the data set into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape , x_test.shape  , x.shape
```

```python
#create model with 5 dense layers
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30,activation='relu',input_shape=(60,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(60,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(30,activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1,activation='sigmoid')
])

```

```python
#Compile and generate the model
model.compile(
    loss = tf.keras.losses.BinaryCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.005), # Fixed the typo here
    metrics = ['accuracy']
)
```

```python
# Train the model
epochs = 100
model.fit(x_train,y_train,epochs=epochs)
```

```python
# Training accuracy = 1.00 that is mean 100% That is good but not the test accuaracy
```

```python
# Evaluate the model
model.evaluate(x_test,y_test)
```
